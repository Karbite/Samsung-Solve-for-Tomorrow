```python
from ipywidgets import HTML
from ipyleaflet import Map, basemaps,basemap_to_tiles, Circle, Marker, Popup
m =Map(center = (latitude, longitude), zoom = 2, min_zoom = 10, max_zoom = 30, basemap = basemaps.OpenStreetMap.France)
# draws circle where hospital data will be acquired
circle = Circle()
circle.location = (latitude, longitude)
circle.radius = 16093
circle.color = "red"
circle.fill_color = "red"
circle.opacity = 1
circle.stroke = False
m.add_layer(circle)
```
